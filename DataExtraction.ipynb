```python
import os
import requests
import json

import pandas as pd
print("pandas: %s" % pd.__version__)
```

```python
# Get all the raw data
response = requests.get("https://min-api.cryptocompare.com/data/all/coinlist")
data = json.loads(response.text)["Data"]
```

```python
# Convert raw data into a dataframe and pick the columns
df = pd.DataFrame.from_dict(
    data, 
    orient="index",
    columns=[
        "CoinName",
        "Algorithm",
        "IsTrading",
        "ProofType",
        "TotalCoinsMined",
        "MaxSupply"
    ]
)
df
```

```python
# Rename column to work with legacy systems
df = df.rename(columns={"MaxSupply": "TotalCoinSupply"})
df
```

```python
if not os.path.exists("Resources"):
    os.makedirs("Resources")
df.to_csv("Resources/crypto_raw_data.csv", index=False)
```
